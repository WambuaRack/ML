```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense ,Activation
import pandas as pd 
from sklearn.model_selection import train_test_split


```

```python
df =pd.read_csv('dataset.csv')
df
```

```python

df['Target']=(df ['Target']== "Dropout").astype(int)
df
```

```python
X_train =df.iloc[:,:-1]
y_train=df.iloc[:,-1]
X_train,X_test,y_train,y_test =train_test_split(X_train,y_train ,test_size=0.4,random_state=42)
X_test,X_val,y_test,y_val =train_test_split(X_test,y_test, test_size=0.2,random_state=42)
```

```python
model = Sequential()
model.add(Dense(32,input_dim =34, activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

df.shape
```

```python
model.summary()
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
```

```python
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train,y_train ,batch_size=64, epochs=16, validation_data=(X_val,y_val))
```

```python
results=model.evaluate(X_test,y_test)
for i in range(len(model.metrics_names)):
    print(model.metrics_names[i], ":", results[i])
```
